````python
import pandas as pd
import requests
from io import BytesIO
from tqdm import tqdm

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/122.0.0.0 Safari/537.36"
    )
}

dataframes = []

for url in tqdm(urls):

    try:

        response = requests.get(
            url,
            headers=headers,
            allow_redirects=True,
            timeout=60
        )

        content = response.content

        filename = url.split("/")[-1]

        if filename.endswith(".xlsx"):

            df = pd.read_excel(
                BytesIO(content),
                engine="openpyxl"
            )

        elif filename.endswith(".xls"):

            df = pd.read_excel(
                BytesIO(content),
                engine="xlrd"
            )

        else:
            print(f"Skipped: {url}")
            continue

        year = None

        if "FY" in filename:
            year = filename.split("FY")[1][:4]

        elif "FMR" in filename:
            year = filename.split("FMR")[1][:4]

        if year and year.isdigit():
            df["year"] = int(year)

        dataframes.append(df)

        print(f"Loaded: {filename} -> {df.shape}")

    except Exception as e:

        print(f"Failed: {url} -> {e}")
````

````python
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd

def load_problematic_xlsx(path):

    with zipfile.ZipFile(path) as z:

        shared_strings = []

        if "xl/sharedStrings.xml" in z.namelist():

            root = ET.fromstring(z.read("xl/sharedStrings.xml"))

            ns = {
                "main":
                "http://schemas.openxmlformats.org/spreadsheetml/2006/main"
            }

            for si in root.findall(".//main:t", ns):
                shared_strings.append(si.text)

        sheet = "xl/worksheets/sheet1.xml"

        root = ET.fromstring(z.read(sheet))

        ns = {
            "main":
            "http://schemas.openxmlformats.org/spreadsheetml/2006/main"
        }

        rows = []

        for row in root.findall(".//main:row", ns):

            vals = []

            for c in row.findall("main:c", ns):

                v = c.find("main:v", ns)

                if v is not None:

                    val = v.text

                    if c.attrib.get("t") == "s":
                        val = shared_strings[int(val)]

                    vals.append(val)

                else:
                    vals.append(None)

            rows.append(vals)

    df = pd.DataFrame(rows)

    df.columns = df.iloc[0]

    df = df[1:].reset_index(drop=True)

    return df
````

````python
# Setting up Data Cleaning Function
def clean_func(df):
    df = df.copy()

    # Standardizing the column names
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
    )

    # Renaming inconsistent columns names
    rename_map = {
        "rent50_0": "rent_50_0",
        "rent50_1": "rent_50_1",
        "rent50_2": "rent_50_2",
        "rent50_3": "rent_50_3",
        "rent50_4": "rent_50_4",

        "rent0": "rent_50_0",
        "rent1": "rent_50_1",
        "rent2": "rent_50_2",
        "rent3": "rent_50_3",
        "rent4": "rent_50_4",

        "rent0br": "rent_50_0",
        "rent1br": "rent_50_1",
        "rent2br": "rent_50_2",
        "rent3br": "rent_50_3",
        "rent4br": "rent_50_4",

        "rent_0": "rent_50_0",
        "rent_1": "rent_50_1",
        "rent_2": "rent_50_2",
        "rent_3": "rent_50_3",
        "rent_4": "rent_50_4",

        "rent50_0bed": "rent_50_0",
        "rent50_1bed": "rent_50_1",
        "rent50_2bed": "rent_50_2",
        "rent50_3bed": "rent_50_3",
        "rent50_4bed": "rent_50_4",

        "state": "state_code",
        "state_alpha": "state_alpha",
        "st": "state_alpha",

        "countyname": "cntyname",
        "county_town_name": "cntyname",
        "name": "cntyname",

        "areaname": "hud_areaname",
        "areaname21": "hud_areaname",
        "areaname20": "hud_areaname",
        "areaname19": "hud_areaname",
        "msaname": "hud_areaname",

        "fips": "fips",
        "fips2010": "fips",
        "fips2000": "fips",
        "fips2025": "fips",

        "pop": "pop",
        "pop2000": "pop",
        "pop100": "pop",
    }

    df = df.rename(columns=rename_map)

    # Creating state_alpha and state_code if not present in dataframes
    if 'state_alpha' not in df.columns:
      df['state_alpha'] = pd.NA
    if 'state_code' not in df.columns:
      df['state_code'] = pd.NA
    # Ensuring string format type for state_alpha
    df['state_alpha'] = (
        df['state_alpha']
        .astype("string")
        .str.strip()
        .str.upper()
        )

    # Creating state_code - state_alpha mapping
    state_alpha_to_code = {
    'AL': 1, 'AK': 2, 'AZ': 4, 'AR': 5, 'CA': 6,
    'CO': 8, 'CT': 9, 'DE': 10, 'DC': 11, 'FL': 12,
    'GA': 13, 'HI': 15, 'ID': 16, 'IL': 17, 'IN': 18,
    'IA': 19, 'KS': 20, 'KY': 21, 'LA': 22, 'ME': 23,
    'MD': 24, 'MA': 25, 'MI': 26, 'MN': 27, 'MS': 28,
    'MO': 29, 'MT': 30, 'NE': 31, 'NV': 32, 'NH': 33,
    'NJ': 34, 'NM': 35, 'NY': 36, 'NC': 37, 'ND': 38,
    'OH': 39, 'OK': 40, 'OR': 41, 'PA': 42, 'RI': 44,
    'SC': 45, 'SD': 46, 'TN': 47, 'TX': 48, 'UT': 49,
    'VT': 50, 'VA': 51, 'WA': 53, 'WV': 54, 'WI': 55,
    'WY': 56,
    'PR': 72,
    'VI': 78,
    'GU': 66,
    'MP': 69,
    'AS': 60
    }

    state_code_to_alpha = {v: k for k, v in state_alpha_to_code.items()}


    # Fixing missing state_code and state_alpha
    if 'state_code' not in df.columns or df['state_code'].isna().any():
      df['state_code'] = df['state_code'].fillna(
          df['state_alpha'].map(state_alpha_to_code)
          )

    if 'state_alpha' not in df.columns or df['state_alpha'].isna().any():
      df['state_alpha'] = df['state_alpha'].fillna(
          df['state_code'].map(state_code_to_alpha)
          )

    # Confirming columns exist across dataframes
    final_cols = [
        'state_code', 'county_code', 'county_sub_code',
        'cntyname', 'town_name', 'hud_areaname',
        'fips', 'hud_area_code',
        'rent_50_0', 'rent_50_1', 'rent_50_2',
        'rent_50_3', 'rent_50_4',
        'population', 'state_alpha', 'year'
    ]

    for col in final_cols:
        if col not in df.columns:
            df[col] = np.nan

    df = df.loc[:, ~df.columns.duplicated()]

    # Fixing nuemric Data type
    numeric_cols = ['rent_50_0', 'rent_50_1',
        'rent_50_2', 'rent_50_3', 'rent_50_4'
    ]

    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Fixing categorical data types
    str_cols = ['cntyname', 'state_alpha']

    for col in str_cols:
      df[col] = df[col].astype(str)
      df[col] = df[col].map(str).str.strip()
      df[col] = df[col].replace(['None', 'nan', 'NaN'], pd.NA)

    # Agggregating the data by state
    grouped_yearly = df.groupby(
        ['year','state_code', 'state_alpha']
        )[['rent_50_0', 'rent_50_1','rent_50_2', 'rent_50_3', 'rent_50_4']].mean().reset_index()

    return df

# Data Cleaning
cleaned_dataframes = [clean_func(df) for df in combined_list]
````

````python
import pandas as pd
import glob
import os

# Import Function
def import_gdp_files(path_pattern="Table_GDP_PI_*.csv"):

    dataframes = {}

    files = sorted(glob.glob(path_pattern))

    for file in files:

        year = os.path.basename(file).split("_")[-1].split(".")[0]

        df = pd.read_csv(
            file,
            skiprows=3
        )

        dataframes[int(year)] = df

        print(f"Loaded {file} -> {df.shape}")

    return dataframes
````

````python
# Applying Functions for data importation
gdp_data = import_gdp_files()
````

### Data Cleaning

````python
# Cleaning + Wide/Long Function
def clean_merge_gdp(dataframes):

    cleaned = []

    for year, df in dataframes.items():

        df = df[["GeoFIPS", "GeoName", str(year)]].copy()

        # keeping only valid numeric GeoFIPS rows
        df["GeoFIPS"] = pd.to_numeric(
            df["GeoFIPS"],
            errors="coerce"
        )

        df = df[df["GeoFIPS"].notna()].copy()

        # converting GeoFIPS
        df["GeoFIPS"] = df["GeoFIPS"].astype(int)

        # splitting county and state
        split_cols = df["GeoName"].str.split(",", expand=True)

        df["county_name"] = split_cols[0].str.strip()

        df["state_alpha"] = split_cols[1].str.strip()

        # renaming GDP column
        df.rename(columns={str(year): f"GDP_PI_{year}"}, inplace=True)

        cleaned.append(df)

    # initializing wide dataframe
    wide_df = cleaned[0][
        [
            "GeoFIPS",
            "GeoName",
            "county_name",
            "state_alpha",
            cleaned[0].columns[2]
        ]
    ].copy()

    # merging remaining years
    for df in cleaned[1:]:

        gdp_col = [c for c in df.columns if "GDP_PI_" in c][0]

        wide_df = wide_df.merge(
            df[["GeoFIPS", gdp_col]],
            on="GeoFIPS",
            how="outer"
        )

    # long format
    long_df = wide_df.melt(
        id_vars=[
            "GeoFIPS",
            "GeoName",
            "county_name",
            "state_alpha"
        ],
        var_name="year",
        value_name="GDP_PI"
    )

    # extracting year
    long_df["year"] = (
        long_df["year"]
        .str.replace("GDP_PI_", "", regex=False)
        .astype(int)
    )

    # numeric GDP
    long_df["GDP_PI"] = pd.to_numeric(
        long_df["GDP_PI"],
        errors="coerce"
    )

    return wide_df, long_df
````

````python
cols = ["GDP_PI", "housing_units", "personal_income"]

merged_df_final = merged_df_final.sort_values(["county_name", "year"])

merged_df_final[cols] = (
    merged_df_final
    .groupby("county_name")[cols]
    .transform(lambda x: x.bfill())
)
````

````python
#Imputing for the two years
cols = ["GDP_PI", "housing_units", "personal_income"]

merged_df_final = merged_df_final.sort_values(
    ["county_name", "year"]
).copy()

# county-wise backward fill
for col in cols:
    merged_df_final[col] = (
        merged_df_final
        .groupby("county_name")[col]
        .transform(lambda x: x.bfill())
    )

# fallback global median
merged_df_final[cols] = (
    merged_df_final[cols]
    .fillna(merged_df_final[cols].median())
)
````

````python
# Rechecking missing entries
merged_df_final.isnull().sum()
````

````python
# Converting data to long
rent_cols = [
    "rent_50_0",
    "rent_50_1",
    "rent_50_2",
    "rent_50_3",
    "rent_50_4"
]

merged_df_final_long = merged_df_final.melt(
    id_vars=[
        "county_name",
        "year",
        "GDP_PI",
        "housing_units",
        "personal_income"
    ],
    value_vars=rent_cols,
    var_name="house_size",
    value_name="rent"
)

# Extracting house size
merged_df_final_long["house_size"] = (
    merged_df_final_long["house_size"]
    .str.rsplit("_", n=1)
    .str[-1]
    .astype(int)
)

# Getting Rent Growth
merged_df_final_long = merged_df_final_long.sort_values(["county_name", "house_size", "year"])

merged_df_final_long["rent_growth"] = (
    merged_df_final_long.groupby(["county_name", "house_size"])["rent"]
    .pct_change()
)
merged_df_final_long
````
